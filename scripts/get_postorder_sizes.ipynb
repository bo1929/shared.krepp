```python
import treeswift as tw
```

```python
t = tw.read_tree("../misc_data/backbone-WoLv1.wnk", schema="newick")
lbl_to_nd = t.label_to_node(selection="all")
```

```python
sizes = ""
visited = set()
with open("../results/clade_versus_multitree_sizes.tsv", 'r') as f:
    for l in f.readlines()[1:]:
        x = l.strip().split()
        nd = lbl_to_nd[(x[0])]
        ts = t.extract_subtree(nd)
        # nd = nd.get_parent().get_parent()
        for ndc in nd.traverse_leaves():
            visited.add(ndc)
        if nd.get_label() == "N1640":
            for ndc in nd.traverse_leaves():
                print(ndc.get_label())
        sizes += (f"{x[0]}\t{len(visited)}\n")
        # print(x[0], len(visited))
```
